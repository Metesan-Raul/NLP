```python
from google.colab import files
uploaded = files.upload()
data = "fra_mixed_2009_10K-words.txt"
languagewords=uploaded[data].decode("utf-8").split("\r\n")
```

```python
!pip install num2words
!pip install unidecode
import unidecode
import re
from num2words import num2words
import emoji
import nltk
nltk.download('punkt')
from nltk import word_tokenize
import string
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python

def preprocess(languagewords):
  language=[]
  for languageword in languagewords:
    languageword=languageword.lower()
    languageword  = unidecode.unidecode(languageword)
    languageword = ''.join([i for i in languageword if not i.isdigit()])
    languageword = re.sub(r'http\S+','', languageword)
    languageword=re.sub(' \d+', '', languageword)
    emoji.get_emoji_regexp().sub(u'', languageword)
    languageword = ' '.join([num2words(word) if word.isdigit() else word for word in languageword.split()])
    languageword = re.sub(r'[^\w\s]','',languageword)
    languageword = languageword.translate(str.maketrans('', '', string.punctuation))
    languageword = re.sub(' +', ' ', languageword)
    tokenized_languageword_nltk = word_tokenize(languageword)
    stop_words_nltk = set(stopwords.words('english'))
    tokenized_languageword_without_stopwords = [i for i in tokenized_languageword_nltk if not i in stop_words_nltk]
    ps = PorterStemmer()
    for word in tokenized_languageword_without_stopwords:
     language.append(ps.stem(word))
  
  return language
preprocess(languagewords)
```
